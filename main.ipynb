**GRU - Gated Recurrent Unit**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score, mean_squared_log_error, r2_score
from datetime import datetime, timedelta

# Load your dataset without parsing dates
file_path = '/content/cleaned_data.csv'
df = pd.read_csv(file_path)

# Print the column names to see what's available
print(df.columns)

# Create a pseudo-date index from 'Year' and 'Week'
df['Date'] = pd.to_datetime(df['Year'].astype(str) + df['Week'].astype(str) + '1', format='%Y%W%w')
df.set_index('Date', inplace=True)

# Now df is indexed by these pseudo-dates
print(df.head())

# Check for missing values
print(df.isna().sum())

# Interpolate missing values
df.interpolate(inplace=True)

# Plot the target variable to understand its pattern
df['Real_Packing'].plot(title='Real Packing over Time', figsize=(14, 7))
plt.show()

# Select the features and the target variables
features = df.drop(['Real_Packing', 'Prod_real_Ha', 'Forecast_Company'], axis=1)
target_real_packing = df[['Real_Packing']]
target_prod_real_ha = df[['Prod_real_Ha']]
target_forecast_company = df[['Forecast_Company']]

# Normalize the features and the target variables
scaler_features = MinMaxScaler()
features_scaled = scaler_features.fit_transform(features)

scaler_target_real_packing = MinMaxScaler()
target_real_packing_scaled = scaler_target_real_packing.fit_transform(target_real_packing)

scaler_target_prod_real_ha = MinMaxScaler()
target_prod_real_ha_scaled = scaler_target_prod_real_ha.fit_transform(target_prod_real_ha)

scaler_target_forecast_company = MinMaxScaler()
target_forecast_company_scaled = scaler_target_forecast_company.fit_transform(target_forecast_company)

# Combine scaled features and targets back into a single DataFrame
df_scaled = pd.DataFrame(features_scaled, index=df.index, columns=features.columns)
df_scaled['Real_Packing'] = target_real_packing_scaled
df_scaled['Prod_real_Ha'] = target_prod_real_ha_scaled
df_scaled['Forecast_Company'] = target_forecast_company_scaled

def create_sequences(data, target, look_back=30):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:(i + look_back)].values)
        y.append(target[i + look_back])
    return np.array(X), np.array(y)

look_back = 30  # Use 30 days of data to predict the next day
X_real_packing, y_real_packing = create_sequences(df_scaled, df_scaled['Real_Packing'], look_back)
X_prod_real_ha, y_prod_real_ha = create_sequences(df_scaled, df_scaled['Prod_real_Ha'], look_back)
X_forecast_company, y_forecast_company = create_sequences(df_scaled, df_scaled['Forecast_Company'], look_back)

# Split the data into training and testing sets
split_point_real_packing = int(len(X_real_packing) * 0.8)
X_train_real_packing, X_test_real_packing = X_real_packing[:split_point_real_packing], X_real_packing[split_point_real_packing:]
y_train_real_packing, y_test_real_packing = y_real_packing[:split_point_real_packing], y_real_packing[split_point_real_packing:]

split_point_prod_real_ha = int(len(X_prod_real_ha) * 0.8)
X_train_prod_real_ha, X_test_prod_real_ha = X_prod_real_ha[:split_point_prod_real_ha], X_prod_real_ha[split_point_prod_real_ha:]
y_train_prod_real_ha, y_test_prod_real_ha = y_prod_real_ha[:split_point_prod_real_ha], y_prod_real_ha[split_point_prod_real_ha:]

split_point_forecast_company = int(len(X_forecast_company) * 0.8)
X_train_forecast_company, X_test_forecast_company = X_forecast_company[:split_point_forecast_company], X_forecast_company[split_point_forecast_company:]
y_train_forecast_company, y_test_forecast_company = y_forecast_company[:split_point_forecast_company], y_forecast_company[split_point_forecast_company:]

# Build the GRU model
def build_and_train_model(X_train, y_train, X_test, y_test, look_back):
    model = Sequential([
        GRU(50, input_shape=(look_back, X_train.shape[2]), return_sequences=True),
        Dropout(0.2),
        GRU(50),
        Dropout(0.2),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')

    # Train the model
    history = model.fit(
        X_train, y_train, epochs=100, batch_size=32,
        validation_data=(X_test, y_test),
        callbacks=[EarlyStopping(monitor='val_loss', patience=10)],
        verbose=1
    )

    # Plot training and validation loss
    plt.figure(figsize=(10, 6))
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.legend()
    plt.show()

    return model

model_real_packing = build_and_train_model(X_train_real_packing, y_train_real_packing, X_test_real_packing, y_test_real_packing, look_back)
model_prod_real_ha = build_and_train_model(X_train_prod_real_ha, y_train_prod_real_ha, X_test_prod_real_ha, y_test_prod_real_ha, look_back)
model_forecast_company = build_and_train_model(X_train_forecast_company, y_train_forecast_company, X_test_forecast_company, y_test_forecast_company, look_back)

# Make predictions
y_pred_real_packing = model_real_packing.predict(X_test_real_packing)
y_pred_prod_real_ha = model_prod_real_ha.predict(X_test_prod_real_ha)
y_pred_forecast_company = model_forecast_company.predict(X_test_forecast_company)

# Inverse transform to get predictions back to the original scale
y_pred_real_packing_inverse = scaler_target_real_packing.inverse_transform(y_pred_real_packing)
y_test_real_packing_inverse = scaler_target_real_packing.inverse_transform(y_test_real_packing.reshape(-1, 1))

y_pred_prod_real_ha_inverse = scaler_target_prod_real_ha.inverse_transform(y_pred_prod_real_ha)
y_test_prod_real_ha_inverse = scaler_target_prod_real_ha.inverse_transform(y_test_prod_real_ha.reshape(-1, 1))

y_pred_forecast_company_inverse = scaler_target_forecast_company.inverse_transform(y_pred_forecast_company)
y_test_forecast_company_inverse = scaler_target_forecast_company.inverse_transform(y_test_forecast_company.reshape(-1, 1))

# Calculate performance metrics for Real_Packing
rmse_real_packing = np.sqrt(mean_squared_error(y_test_real_packing_inverse, y_pred_real_packing_inverse))
mae_real_packing = mean_absolute_error(y_test_real_packing_inverse, y_pred_real_packing_inverse)
evs_real_packing = explained_variance_score(y_test_real_packing_inverse, y_pred_real_packing_inverse)
msle_real_packing = mean_squared_log_error(y_test_real_packing_inverse, y_pred_real_packing_inverse)
r2_real_packing = r2_score(y_test_real_packing_inverse, y_pred_real_packing_inverse)

print(f'Real_Packing - RMSE: {rmse_real_packing}')
print(f'Real_Packing - MAE: {mae_real_packing}')
print(f'Real_Packing - EVS: {evs_real_packing}')
print(f'Real_Packing - MSLE: {msle_real_packing}')
print(f'Real_Packing - R2: {r2_real_packing}')

# Calculate performance metrics for Prod_real_Ha
rmse_prod_real_ha = np.sqrt(mean_squared_error(y_test_prod_real_ha_inverse, y_pred_prod_real_ha_inverse))
mae_prod_real_ha = mean_absolute_error(y_test_prod_real_ha_inverse, y_pred_prod_real_ha_inverse)
evs_prod_real_ha = explained_variance_score(y_test_prod_real_ha_inverse, y_pred_prod_real_ha_inverse)
msle_prod_real_ha = mean_squared_log_error(y_test_prod_real_ha_inverse, y_pred_prod_real_ha_inverse)
r2_prod_real_ha = r2_score(y_test_prod_real_ha_inverse, y_pred_prod_real_ha_inverse)

print(f'Prod_real_Ha - RMSE: {rmse_prod_real_ha}')
print(f'Prod_real_Ha - MAE: {mae_prod_real_ha}')
print(f'Prod_real_Ha - EVS: {evs_prod_real_ha}')
print(f'Prod_real_Ha - MSLE: {msle_prod_real_ha}')
print(f'Prod_real_Ha - R2: {r2_prod_real_ha}')

# Calculate performance metrics for Forecast_Company
rmse_forecast_company = np.sqrt(mean_squared_error(y_test_forecast_company_inverse, y_pred_forecast_company_inverse))
mae_forecast_company = mean_absolute_error(y_test_forecast_company_inverse, y_pred_forecast_company_inverse)
evs_forecast_company = explained_variance_score(y_test_forecast_company_inverse, y_pred_forecast_company_inverse)
msle_forecast_company = mean_squared_log_error(y_test_forecast_company_inverse, y_pred_forecast_company_inverse)
r2_forecast_company = r2_score(y_test_forecast_company_inverse, y_pred_forecast_company_inverse)

print(f'Forecast_Company - RMSE: {rmse_forecast_company}')
print(f'Forecast_Company - MAE: {mae_forecast_company}')
print(f'Forecast_Company - EVS: {evs_forecast_company}')
print(f'Forecast_Company - MSLE: {msle_forecast_company}')
print(f'Forecast_Company - R2: {r2_forecast_company}')

# Plot actual vs predicted for Real_Packing
plt.figure(figsize=(15, 8))
plt.plot(y_test_real_packing_inverse, label='Actual Real_Packing')
plt.plot(y_pred_real_packing_inverse, label='Predicted Real_Packing')
plt.legend()
plt.title('Real Packing Forecast vs Actuals')
plt.show()

# Plot actual vs predicted for Prod_real_Ha
plt.figure(figsize=(15, 8))
plt.plot(y_test_prod_real_ha_inverse, label='Actual Prod_real_Ha')
plt.plot(y_pred_prod_real_ha_inverse, label='Predicted Prod_real_Ha')
plt.legend()
plt.title('Prod_real_Ha Forecast vs Actuals')
plt.show()

# Plot actual vs predicted for Forecast_Company
plt.figure(figsize=(15, 8))
plt.plot(y_test_forecast_company_inverse, label='Actual Forecast_Company')
plt.plot(y_pred_forecast_company_inverse, label='Predicted Forecast_Company')
plt.legend()
plt.title('Forecast_Company Forecast vs Actuals')
plt.show()

# Generate synthetic data for 2018
np.random.seed(42)
start_date = datetime(2018, 1, 1)
end_date = datetime(2018, 12, 31)
date_generated = [start_date + timedelta(days=x) for x in range(0, (end_date-start_date).days + 1, 7)]
week_numbers = [date.isocalendar()[1] for date in date_generated]

synthetic_data = {
    'Year': [2018] * len(date_generated),
    'Week': week_numbers,
    'Real_Packing': np.random.uniform(low=100, high=200, size=len(date_generated)),
    'Prod_real_Ha': np.random.uniform(low=50, high=100, size=len(date_generated)),
    'Forecast_Company': np.random.uniform(low=200, high=300, size=len(date_generated))
}

synthetic_df = pd.DataFrame(synthetic_data)
synthetic_df['Date'] = date_generated

# Plot synthetic data for Real_Packing, Prod_real_Ha, and Forecast_Company
plt.figure(figsize=(15, 5))
plt.plot(synthetic_df['Date'], synthetic_df['Real_Packing'], label='Real_Packing')
plt.plot(synthetic_df['Date'], synthetic_df['Prod_real_Ha'], label='Prod_real_Ha')
plt.plot(synthetic_df['Date'], synthetic_df['Forecast_Company'], label='Forecast_Company')
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Synthetic Data for 2018')
plt.legend()
plt.grid(True)
plt.show()
```

**Prophet**

```python
import pandas as pd
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score, mean_squared_log_error, r2_score

# Load the dataset
file_path = '/content/cleaned_data.csv'
data = pd.read_csv(file_path)

# Function to convert Year and Week to a datetime format
def year_week_to_date(year, week):
    first_day_of_year = datetime(int(year), 1, 1)
    if first_day_of_year.weekday() > 3:
        first_day_of_year += timedelta(7 - first_day_of_year.weekday())
    else:
        first_day_of_year -= timedelta(first_day_of_year.weekday())
    return first_day_of_year + timedelta(days=int(week) * 7)

# Ensure 'Year' and 'Week' columns are integers
data['Year'] = data['Year'].astype(int)
data['Week'] = data['Week'].astype(int)

# Generate the 'ds' column using the lists
data['ds'] = data.apply(lambda row: year_week_to_date(row['Year'], row['Week']), axis=1)

# Function to fit and evaluate the Prophet model
def fit_and_evaluate_prophet_model(df, target_col):
    df = df.rename(columns={'ds': 'ds', target_col: 'y'})
    model = Prophet()
    model.fit(df[['ds', 'y']])

    # Cross-validation
    df_cv = cross_validation(model, initial='730 days', period='180 days', horizon='365 days')
    df_performance = performance_metrics(df_cv)

    return model, df_cv, df_performance

# Fit models and evaluate for 'Prod_real_Ha', 'Forecast_Company', and 'Real_Packing'
results = {}
for target in ['Prod_real_Ha', 'Forecast_Company', 'Real_Packing']:
    df = data[['ds', target]].dropna().rename(columns={target: 'y'})
    model, df_cv, df_performance = fit_and_evaluate_prophet_model(df, 'y')
    results[target] = {
        'model': model,
        'cross_validation': df_cv,
        'performance': df_performance
    }

# Generate synthetic data for 2018
future_dates = pd.date_range(start='2018-01-01', end='2018-12-31', freq='W')
synthetic_data = pd.DataFrame({'ds': future_dates})

# Plot the forecasts and performance metrics, calculate additional metrics
for target, result in results.items():
    model = result['model']
    forecast = model.predict(synthetic_data)

    # Generate actual synthetic target values (e.g., random values)
    synthetic_data['y'] = forecast['yhat'] + (forecast['yhat'] * 0.1)  # adding 10% noise for synthetic actuals

    # Calculate the evaluation metrics
    mae = mean_absolute_error(synthetic_data['y'], forecast['yhat'])
    mse = mean_squared_error(synthetic_data['y'], forecast['yhat'])
    evs = explained_variance_score(synthetic_data['y'], forecast['yhat'])
    msle = mean_squared_log_error(synthetic_data['y'], forecast['yhat'])
    r2 = r2_score(synthetic_data['y'], forecast['yhat'])

    print(f"Evaluation metrics for {target}:")
    print(f"Mean Absolute Error: {mae}")
    print(f"Mean Squared Error: {mse}")
    print(f"Explained Variance Score: {evs}")
    print(f"Mean Squared Log Error: {msle}")
    print(f"R2 Score: {r2}")

    # Plot the forecast
    fig, ax = plt.subplots(figsize=(10, 6))
    model.plot(forecast, ax=ax)
    ax.set_title(f"Forecast for {target} for 2018")
    plt.show()
```

**Hybrid Models - RNN,random Forest,XGBoost**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score, mean_squared_log_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from datetime import datetime, timedelta

# Load your dataset
data = pd.read_csv('/content/cleaned_data.csv')

# Prepare the data
data['date'] = pd.to_datetime(data['Year'].astype(str) + data['Week'].astype(str).str.zfill(2) + '1', format='%Y%W%w')
data.set_index('date', inplace=True)

# Select the features and target variable
features = ['Real_Packing', 'Prod_real_Ha', 'Forecast_Company']  # Add other feature columns as needed
data = data[features]

# Normalize the data
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)

# Create sequences for RNN
def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i + sequence_length])
        y.append(data[i + sequence_length, :])  # Predicting all target variables
    return np.array(X), np.array(y)

sequence_length = 12
X, y = create_sequences(data_scaled, sequence_length)

# Split the data
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Train RNN model
rnn_model = Sequential()
rnn_model.add(SimpleRNN(50, return_sequences=True, input_shape=(sequence_length, len(features))))
rnn_model.add(SimpleRNN(50))
rnn_model.add(Dense(len(features)))  # Predicting all target variables
rnn_model.compile(optimizer='adam', loss='mean_squared_error')
history = rnn_model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2)

# Predict using RNN
y_train_pred_rnn = rnn_model.predict(X_train)
y_test_pred_rnn = rnn_model.predict(X_test)

# Plot training and validation loss
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Calculate residuals
train_residuals = y_train - y_train_pred_rnn
test_residuals = y_test - y_test_pred_rnn

# Prepare data for Random Forest and XGBoost with additional features
lags = 12

def create_lag_features(data, lags):
    lag_features = pd.DataFrame()
    for lag in range(1, lags + 1):
        for i, col in enumerate(data.columns):
            lag_features[f'lag_{lag}_{col}'] = data[col].shift(lag)
    return lag_features.dropna()

data_df = pd.DataFrame(data_scaled, columns=features)
train_residuals_df = create_lag_features(data_df.iloc[:len(train_residuals)], lags)
test_residuals_df = create_lag_features(data_df.iloc[len(train_residuals):], lags)

# Align the residuals with the lags
aligned_train_residuals = train_residuals[-len(train_residuals_df):]
aligned_test_residuals = test_residuals[-len(test_residuals_df):]

# Train Random Forest on residuals
rf_model = RandomForestRegressor(n_estimators=100)
rf_model.fit(train_residuals_df, aligned_train_residuals)

# Predict using Random Forest
train_residuals_pred_rf = rf_model.predict(train_residuals_df)
test_residuals_pred_rf = rf_model.predict(test_residuals_df)

# Calculate new residuals
train_residuals_rf = aligned_train_residuals - train_residuals_pred_rf
test_residuals_rf = aligned_test_residuals - test_residuals_pred_rf

# Prepare data for XGBoost
train_residuals_rf_df = create_lag_features(data_df.iloc[:len(train_residuals_rf)], lags)
test_residuals_rf_df = create_lag_features(data_df.iloc[len(train_residuals_rf):], lags)

# Align the residuals with the lags
aligned_train_residuals_rf = train_residuals_rf[-len(train_residuals_rf_df):]
aligned_test_residuals_rf = test_residuals_rf[-len(test_residuals_rf_df):]

# Train XGBoost on residuals
xgb_model = XGBRegressor(n_estimators=100)
xgb_model.fit(train_residuals_rf_df, aligned_train_residuals_rf)

# Predict using XGBoost
train_residuals_pred_xgb = xgb_model.predict(train_residuals_rf_df)
test_residuals_pred_xgb = xgb_model.predict(test_residuals_rf_df)

# Combine predictions
min_length_train = min(len(y_train_pred_rnn[lags:]), len(train_residuals_pred_rf), len(train_residuals_pred_xgb))
final_train_pred = y_train_pred_rnn[lags:][:min_length_train] + train_residuals_pred_rf[:min_length_train].reshape(-1, len(features)) + train_residuals_pred_xgb[:min_length_train].reshape(-1, len(features))

min_length_test = min(len(y_test_pred_rnn[lags:]), len(test_residuals_pred_rf), len(test_residuals_pred_xgb))
final_test_pred = y_test_pred_rnn[lags:][:min_length_test] + test_residuals_pred_rf[:min_length_test].reshape(-1, len(features)) + test_residuals_pred_xgb[:min_length_test].reshape(-1, len(features))

# Inverse transform predictions
final_train_pred_inverse = scaler.inverse_transform(final_train_pred)
final_test_pred_inverse = scaler.inverse_transform(final_test_pred)
y_test_inverse = scaler.inverse_transform(y_test[lags:][:min_length_test])

# Calculate mean absolute error
mae_real_packing = mean_absolute_error(y_test_inverse[:, 0], final_test_pred_inverse[:, 0])
mae_prod_real_ha = mean_absolute_error(y_test_inverse[:, 1], final_test_pred_inverse[:, 1])
mae_forecast_company = mean_absolute_error(y_test_inverse[:, 2], final_test_pred_inverse[:, 2])
print("Mean Absolute Error for Real_Packing:", mae_real_packing)
print("Mean Absolute Error for Prod_real_Ha:", mae_prod_real_ha)
print("Mean Absolute Error for Forecast_Company:", mae_forecast_company)

# Plot predictions vs actual for Real_Packing
plt.figure(figsize=(15, 5))
plt.plot(y_test_inverse[:, 0], label='Actual Real_Packing')
plt.plot(final_test_pred_inverse[:, 0], label='Predicted Real_Packing')
plt.xlabel('Date')
plt.ylabel('Real Packing')
plt.legend()
plt.show()

# Plot actual and predicted Prod_real_Ha
plt.figure(figsize=(15, 5))
plt.plot(y_test_inverse[:, 1], label='Actual Prod_real_Ha')
plt.plot(final_test_pred_inverse[:, 1], label='Predicted Prod_real_Ha')
plt.xlabel('Date')
plt.ylabel('Prod_real_Ha')
plt.legend()
plt.show()

# Plot actual and predicted Forecast_Company
plt.figure(figsize=(15, 5))
plt.plot(y_test_inverse[:, 2], label='Actual Forecast_Company')
plt.plot(final_test_pred_inverse[:, 2], label='Predicted Forecast_Company')
plt.xlabel('Date')
plt.ylabel('Forecast_Company')
plt.legend()
plt.show()

# Calculate performance metrics
rmse_real_packing = np.sqrt(mean_squared_error(y_test_inverse[:, 0], final_test_pred_inverse[:, 0]))
mae_real_packing = mean_absolute_error(y_test_inverse[:, 0], final_test_pred_inverse[:, 0])
evs_real_packing = explained_variance_score(y_test_inverse[:, 0], final_test_pred_inverse[:, 0])
msle_real_packing = mean_squared_log_error(y_test_inverse[:, 0], final_test_pred_inverse[:, 0])
r2_real_packing = r2_score(y_test_inverse[:, 0], final_test_pred_inverse[:, 0])

rmse_prod_real_ha = np.sqrt(mean_squared_error(y_test_inverse[:, 1], final_test_pred_inverse[:, 1]))
mae_prod_real_ha = mean_absolute_error(y_test_inverse[:, 1], final_test_pred_inverse[:, 1])
evs_prod_real_ha = explained_variance_score(y_test_inverse[:, 1], final_test_pred_inverse[:, 1])
msle_prod_real_ha = mean_squared_log_error(y_test_inverse[:, 1], final_test_pred_inverse[:, 1])
r2_prod_real_ha = r2_score(y_test_inverse[:, 1], final_test_pred_inverse[:, 1])

rmse_forecast_company = np.sqrt(mean_squared_error(y_test_inverse[:, 2], final_test_pred_inverse[:, 2]))
mae_forecast_company = mean_absolute_error(y_test_inverse[:, 2], final_test_pred_inverse[:, 2])
evs_forecast_company = explained_variance_score(y_test_inverse[:, 2], final_test_pred_inverse[:, 2])
msle_forecast_company = mean_squared_log_error(y_test_inverse[:, 2], final_test_pred_inverse[:, 2])
r2_forecast_company = r2_score(y_test_inverse[:, 2], final_test_pred_inverse[:, 2])

# Print the metrics
print("Metrics for Real_Packing:")
print(f"RMSE: {rmse_real_packing}")
print(f"MAE: {mae_real_packing}")
print(f"Explained Variance Score: {evs_real_packing}")
print(f"Mean Squared Log Error: {msle_real_packing}")
print(f"R2 Score: {r2_real_packing}")

print("\nMetrics for Prod_real_Ha:")
print(f"RMSE: {rmse_prod_real_ha}")
print(f"MAE: {mae_prod_real_ha}")
print(f"Explained Variance Score: {evs_prod_real_ha}")
print(f"Mean Squared Log Error: {msle_prod_real_ha}")
print(f"R2 Score: {r2_prod_real_ha}")

print("\nMetrics for Forecast_Company:")
print(f"RMSE: {rmse_forecast_company}")
print(f"MAE: {mae_forecast_company}")
print(f"Explained Variance Score: {evs_forecast_company}")
print(f"Mean Squared Log Error: {msle_forecast_company}")
print(f"R2 Score: {r2_forecast_company}")

# Generate synthetic data for 2018
np.random.seed(42)
start_date = datetime(2018, 1, 1)
end_date = datetime(2018, 12, 31)
date_generated = [start_date + timedelta(days=x) for x in range(0, (end_date-start_date).days + 1, 7)]
week_numbers = [date.isocalendar()[1] for date in date_generated]

synthetic_data = {
    'Year': [2018] * len(date_generated),
    'Week': week_numbers,
    'Real_Packing': np.random.uniform(low=100, high=200, size=len(date_generated)),
    'Prod_real_Ha': np.random.uniform(low=50, high=100, size=len(date_generated)),
    'Forecast_Company': np.random.uniform(low=200, high=300, size=len(date_generated))
}

synthetic_df = pd.DataFrame(synthetic_data)
synthetic_df['Date'] = date_generated

# Plot synthetic data
plt.figure(figsize=(15, 5))
plt.plot(synthetic_df['Date'], synthetic_df['Real_Packing'], label='Real_Packing')
plt.plot(synthetic_df['Date'], synthetic_df['Prod_real_Ha'], label='Prod_real_Ha')
plt.plot(synthetic_df['Date'], synthetic_df['Forecast_Company'], label='Forecast_Company')
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Synthetic Data for 2018')
plt.legend()
plt.grid(True)
plt.show()
```

**Bidirectional LSTM**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score, mean_squared_log_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional

# Load your dataset
data = pd.read_csv('/content/cleaned_data.csv')

# Prepare the data for LSTM, including the additional feature Prod_real_Ha and Forecast_Company
df_lstm = data[['Year', 'Week', 'Real_Packing', 'Prod_real_Ha', 'Forecast_Company']].copy()
df_lstm['date'] = pd.to_datetime(df_lstm['Year'].astype(str) + df_lstm['Week'].astype(str).str.zfill(2) + '1', format='%Y%W%w')
df_lstm.set_index('date', inplace=True)
df_lstm = df_lstm.drop(['Year', 'Week'], axis=1)

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df_lstm)

# Create sequences of 12 weeks (3 months)
def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i + sequence_length])
        y.append(data[i + sequence_length, :])  # Predicting Real_Packing, Prod_real_Ha, and Forecast_Company
    return np.array(X), np.array(y)

sequence_length = 12
X, y = create_sequences(scaled_data, sequence_length)

# Split the data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build the Bidirectional LSTM model
model = Sequential()
model.add(Bidirectional(LSTM(units=100, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2]))))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(units=100, return_sequences=True)))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(units=50)))
model.add(Dropout(0.2))
model.add(Dense(3))  # Output layer with 3 units for Real_Packing, Prod_real_Ha, and Forecast_Company
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)

# Make predictions
y_pred = model.predict(X_test)
y_pred_inverse = scaler.inverse_transform(np.concatenate((y_pred, np.zeros((y_pred.shape[0], scaled_data.shape[1] - 3))), axis=1))[:, :3]
y_test_inverse = scaler.inverse_transform(np.concatenate((y_test, np.zeros((y_test.shape[0], scaled_data.shape[1] - 3))), axis=1))[:, :3]

# Calculate performance metrics for Real_Packing
rmse_real_packing = np.sqrt(mean_squared_error(y_test_inverse[:, 0], y_pred_inverse[:, 0]))
mae_real_packing = mean_absolute_error(y_test_inverse[:, 0], y_pred_inverse[:, 0])
evs_real_packing = explained_variance_score(y_test_inverse[:, 0], y_pred_inverse[:, 0])
msle_real_packing = mean_squared_log_error(y_test_inverse[:, 0], y_pred_inverse[:, 0])
r2_real_packing = r2_score(y_test_inverse[:, 0], y_pred_inverse[:, 0])

# Calculate performance metrics for Prod_real_Ha
rmse_prod_real_ha = np.sqrt(mean_squared_error(y_test_inverse[:, 1], y_pred_inverse[:, 1]))
mae_prod_real_ha = mean_absolute_error(y_test_inverse[:, 1], y_pred_inverse[:, 1])
evs_prod_real_ha = explained_variance_score(y_test_inverse[:, 1], y_pred_inverse[:, 1])
msle_prod_real_ha = mean_squared_log_error(y_test_inverse[:, 1], y_pred_inverse[:, 1])
r2_prod_real_ha = r2_score(y_test_inverse[:, 1], y_pred_inverse[:, 1])

# Calculate performance metrics for Forecast_Company
rmse_forecast_company = np.sqrt(mean_squared_error(y_test_inverse[:, 2], y_pred_inverse[:, 2]))
mae_forecast_company = mean_absolute_error(y_test_inverse[:, 2], y_pred_inverse[:, 2])
evs_forecast_company = explained_variance_score(y_test_inverse[:, 2], y_pred_inverse[:, 2])
msle_forecast_company = mean_squared_log_error(y_test_inverse[:, 2], y_pred_inverse[:, 2])
r2_forecast_company = r2_score(y_test_inverse[:, 2], y_pred_inverse[:, 2])

# Print the metrics
print("Metrics for Real_Packing:")
print(f"RMSE: {rmse_real_packing}")
print(f"MAE: {mae_real_packing}")
print(f"Explained Variance Score: {evs_real_packing}")
print(f"Mean Squared Log Error: {msle_real_packing}")
print(f"R2 Score: {r2_real_packing}")

print("\nMetrics for Prod_real_Ha:")
print(f"RMSE: {rmse_prod_real_ha}")
print(f"MAE: {mae_prod_real_ha}")
print(f"Explained Variance Score: {evs_prod_real_ha}")
print(f"Mean Squared Log Error: {msle_prod_real_ha}")
print(f"R2 Score: {r2_prod_real_ha}")

print("\nMetrics for Forecast_Company:")
print(f"RMSE: {rmse_forecast_company}")
print(f"MAE: {mae_forecast_company}")
print(f"Explained Variance Score: {evs_forecast_company}")
print(f"Mean Squared Log Error: {msle_forecast_company}")
print(f"R2 Score: {r2_forecast_company}")

# Plot training history
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot predictions vs actual for Real_Packing
plt.figure(figsize=(15, 5))
plt.plot(y_test_inverse[:, 0], label='Actual Real_Packing')
plt.plot(y_pred_inverse[:, 0], label='Predicted Real_Packing')
plt.xlabel('Date')
plt.ylabel('Real Packing')
plt.legend()
plt.show()

# Plot predictions vs actual for Prod_real_Ha
plt.figure(figsize=(15, 5))
plt.plot(y_test_inverse[:, 1], label='Actual Prod_real_Ha')
plt.plot(y_pred_inverse[:, 1], label='Predicted Prod_real_Ha')
plt.xlabel('Date')
plt.ylabel('Prod_real_Ha')
plt.legend()
plt.show()

# Plot predictions vs actual for Forecast_Company
plt.figure(figsize=(15, 5))
plt.plot(y_test_inverse[:, 2], label='Actual Forecast_Company')
plt.plot(y_pred_inverse[:, 2], label='Predicted Forecast_Company')
plt.xlabel('Date')
plt.ylabel('Forecast_Company')
plt.legend()
plt.show()

# Plot Prod_real_Ha, Real_Packing, and Forecast_Company using secondary y-axis
fig, ax1 = plt.subplots(figsize=(15, 5))

color = 'tab:blue'
ax1.set_xlabel('Date')
ax1.set_ylabel('Prod_real_Ha', color=color)
ax1.plot(y_test_inverse[:, 1], color=color, label='Actual Prod_real_Ha')
ax1.plot(y_pred_inverse[:, 1], color='tab:cyan', label='Predicted Prod_real_Ha')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:orange'
ax2.set_ylabel('Real Packing', color=color)  # we already handled the x-label with ax1
ax2.plot(y_test_inverse[:, 0], color=color, label='Actual Real_Packing')
ax2.plot(y_pred_inverse[:, 0], color='tab:green', label='Predicted Real_Packing')
ax2.tick_params(axis='y', labelcolor=color)

ax3 = ax1.twinx()  # instantiate a third axes that shares the same x-axis
ax3.spines["right"].set_position(("outward", 60))
color = 'tab:red'
ax3.set_ylabel('Forecast_Company', color=color)
ax3.plot(y_test_inverse[:, 2], color=color, label='Actual Forecast_Company')
ax3.plot(y_pred_inverse[:, 2], color='tab:purple', label='Predicted Forecast_Company')
ax3.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
plt.show()
```
